# data cleaning

# # restriction 1 
### using only 2017 data

```python
data.head(7)
```

```python
# we only need the data from 2017
data = data[data['Year'] == 2017]
data.head(7)
```

```python
# restriction in this case data from 2017
data = data[data['Year'] == 2017]
data.head(7)
```

```python
# drop the duplicate values of players and keep the last join team in 2017
data.drop_duplicates(subset="Player", keep="last", inplace=True, ignore_index=True)
data.head(7)
```

# # restriction 2
### drop coulmn has many NaN data

```python
data_nn = data.isna().sum().sort_values(ascending=False)
data_nn[data_nn >= 1]
```

```python
#because column blank2 and column blanl many missing values, we must drop it
data = data.drop(columns = ['blank2','blanl'])
data.head(7)
```

```python
data_nn = data.isna().sum().sort_values(ascending=False)
data_nn[data_nn >= 1]
```

```python
#choose column names with null values
null = ['3P%','FT%','2P%','TS%','TOV%','FG%','eFG%','FTr','3PAr']
list(enumerate(null))
```

```python
#plot distribution of column with null values
plt.figure(figsize = (15,30))
for i in enumerate(null):
    plt.subplot(5,3,i[0]+1)
    sns.distplot(data[i[1]])

import warnings
warnings.filterwarnings("ignore")
```

```python
#see the distribution skew, if middle skew nan values replaced by mean, if right or left skew, NaN values replaced by median
data['3P%'] = data['3P%'].fillna(data['3P%'].median())
data['FT%'] = data['FT%'].fillna(data['FT%'].median())
data['2P%'] = data['2P%'].fillna(data['2P%'].mean())
data['TS%'] = data['TS%'].fillna(data['TS%'].median())
data['TOV%'] = data['TOV%'].fillna(data['TOV%'].median())
data['FG%'] = data['FG%'].fillna(data['FG%'].mean())
data['eFG%'] = data['eFG%'].fillna(data['eFG%'].mean())
data['FTr'] = data['FTr'].fillna(data['FT%'].median())
data['3PAr'] = data['3PAr'].fillna(data['3PAr'].median())
```

```python
#check null values again
data_nnn = data.isna().sum().sort_values(ascending=False)
data_nn[data_nnn >= 1]
```

## question 3
### Which team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%)?

```python
col = ['TRB%',"AST%","STL%","BLK%"]
for i in col:
    team = data[['Tm',i]]
    team = team.groupby(["Tm"]).mean().reset_index(False)
    team = team[team[i] == team[i].max()]
    display(team.reset_index())
    print()
```

team has the highest average total rebound percentage (TRB%), assist percentage (AST%), steal percentage (STL%), and block percentage (BLK%) can seen before, by TRB is WAS team, by AST is DEN team, by STL is MIN team and by BLK is MIL

### Who is the best player in your opinion based on his record stats?

top player efficiency = FT + 2*2P + 3*3P + DRB + ORB + AST + STL - 0.5*(FT-FTA) - (2P-2PA) - (3P-3PA) - TOV - PF

```python
d = data
```
